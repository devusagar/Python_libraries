```python
# Reshape x with 2 rows and 8 columns
import numpy as np

# Example x with 16 elements
x = np.arange(1, 17)

# Reshape to 2 rows and 8 columns
x_reshaped = x.reshape(2, 8)
print(x_reshaped)
```

As you can see above that our x changed into 2D matrix

2. Reshaping 1-D to 3-D array

```python
# reshaoe x with dimension that will have 2 arrays that contains 4 arrays, each with 2 elements:
import numpy as np
x = np.arange(1, 17)
x_reshaped=x.reshape(2,4,2)
print(x_reshaped)
```

```python
# Use unknown dimention to reshape x into 2-D numpy array with shape 4*4
import numpy as np
x = np.arange(1, 17)
x_reshaped=x.reshape(4,4)
print(x_reshaped)
```
